```python
from sklearn.datasets import fetch_openml
dataset = fetch_openml('mnist_784')
```

```python
# data and target 

X = dataset.data
y = dataset.target
```

```python
plt.imshow(X[0].reshape(28,28))      # imshow = image show
plt.axis('off')
plt.show()
```

```python
plt.imshow(X[0].reshape(28,28),'binary')
plt.axis('off')
plt.show()
```

```python
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(X[i].reshape(28,28),'binary')
    plt.axis('off')
plt.show()
```

```python
X = X/255     # easy way of scaling all the variables are now betwwen 0 to 1 
```

```python
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(X[i].reshape(28,28),'binary')
    plt.axis('off')
plt.show()
# after scaling
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
%time log_reg.fit(X_train, y_train)    #  %time = to find out the time the code is taking time
```

```python
print('Training Score')
print(log_reg.score(X_train,y_train))
print('Test Score')
print(log_reg.score(X_test,y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
%time knn.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(X_test[i].reshape(28,28),'binary')
    plt.xlabel('Actual : {} \nPredicted : {}' .format(y_test[i],y_pred[i]))
plt.tight_layout()
plt.show()
```
